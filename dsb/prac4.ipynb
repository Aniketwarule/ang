```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the datasets
air_df = pd.read_csv("AirQuality_visualization.csv", delimiter=";")
heart_df = pd.read_csv("heart.csv")
forest_df = pd.read_csv("forestfires.csv")
```

```python
air_df_cleaned = air_df.dropna()
heart_df_cleaned = heart_df.dropna()
forest_df_cleaned = forest_df.dropna()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=forest_df_cleaned, x='month', y='area', estimator=sum, errorbar=None)
plt.title("Total Area Burned per Month")
plt.xlabel("Month")
plt.ylabel("Total Area Burned")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=forest_df_cleaned, x='wind', y='temp', hue='area', size='area')
plt.title("Wind vs Temperature (Forest Fires)")
plt.xlabel("Wind")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(heart_df_cleaned.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Heart Data")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=heart_df_cleaned, x='age', bins=10, kde=True)
plt.title("Age Distribution of Heart Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sizes = heart_df_cleaned['output'].value_counts()
labels = ['No Disease', 'Has Disease']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Heart Disease Distribution")
plt.axis('equal')
plt.show()
```
